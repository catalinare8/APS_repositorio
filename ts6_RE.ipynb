Dadas las siguientes ecuaciones de diferencias de los siguientes sistemas, que representan un filtro de media móvil:
 $$a) y(n)=x(n)=x(n-3)+x(n-2)+x(n-1)+x(n)$$
 $$b) y(n)=x(n-4)+x(n-3)+x(n-2)+x(n-1)+x(n)$$
 $$c) y(n)=x(n)-x(n-1)$$
 $$d) y(n)=x(n)-x(n-2)$$
Se pide: 
- Hallar $T(z)=\frac{Y(z)}{X(z)}$
- Calcular su respeusta en frecuencia de módulo y fase.
- Simular y validar la respuesta en frecuencia de todos los sistemas con Numpy.

a)
$$Y(z)=X(z)z^{-3}+X(z)z^{-2}+X(z)z^{-1}+X(z)$$
$$T(z)=\frac{Y(z)}{X(z)}=z^{-3}+z^{-2}+z^{-1}+1=z^{-3}(1+z+z^2+z^3)$$
$$T(j\Omega)=\frac{1+e^{j\Omega}+e^{j2\Omega}+e^{j3\Omega}}{e^{j3\Omega}}$$
$$T(j\Omega)=\frac{e^{j\frac{3}{2}\Omega}(e^{-j\frac{3}{2}\Omega}+e^{-j\frac{1}{2}\Omega}+e^{j\frac{1}{2}\Omega}+e^{j\frac{3}{2}\Omega})}{e^{j3\Omega}}$$
$$T(j\Omega)=e^{-j\frac{3}{2}\Omega}(2cos(\frac{3}{2}\Omega)+2cos(\frac{1}{2}\Omega)$$
Fase: $e^{-j\frac{3}{2}\Omega}$ 

Módulo: $|(2cos(\frac{3}{2}\Omega)+2cos(\frac{1}{2}\Omega)|$


b)
$$Y(z)=X(z)z^{-4}+X(z)z^{-3}+X(z)z^{-2}+X(z)z^{-1}+X(z)$$
$$T(z)=\frac{Y(z)}{X(z)}=z^{-4}+z^{-3}+z^{-2}+z^{-1}+1=z^{-4}(1+z+z^2+z^3+z^4)$$
$$T(j\Omega)=\frac{1+e^{j\Omega}+e^{j2\Omega}+e^{j3\Omega}+e^{j4\Omega}}{e^{j4\Omega}}$$
$$T(j\Omega)=\frac{e^{j2\Omega}(e^{-j2\Omega}+e^{-j1\Omega}+1+e^{j1\Omega}+e^{j2\Omega}}{e^{j4\Omega}}$$
$$T(j\Omega)=e^{-j2\Omega}(e^{-j2\Omega}+e^{-j1\Omega}+1+e^{j1\Omega}+e^{j2\Omega})$$
$$T(j\Omega)=e^{-j2\Omega}(1+2cos(2\Omega)+2cos(\Omega))$$
Fase: $e^{-j2\Omega}$ 

Módulo: $|(1+2cos(2\Omega)+2cos(\Omega))|$

c)
$$Y(z)=X(z)-X(z)z^{-1}$$
$$T(z)=\frac{Y(z)}{X(z)}=1-z^{-1}$$
$$T(j\Omega)=1-e^{-j1\Omega}$$
$$T(j\Omega)=e^{-j\frac{\Omega}{2}}(e^{j\frac{\Omega}{2}}-e^{-j\frac{\Omega}{2}})$$
$$T(j\Omega)=e^{-j\frac{\Omega}{2}}2jsen(\frac{\Omega}{2})$$
$$T(j\Omega)=e^{-j(\frac{\Omega}{2}-\frac{\pi}{2})}2sen(\frac{\Omega}{2})$$
Fase: $e^{-j(\frac{\Omega}{2}-\frac{\pi}{2})}$

Módulo: $|2sen(\frac{\Omega}{2})|$

d)
$$Y(z)=X(z)-X(z)z^{-2}$$
$$T(z)=\frac{Y(z)}{X(z)}=1-z^{-2}$$
$$T(j\Omega)=1-e^{-j2\Omega}$$
$$T(j\Omega)=e^{-j\Omega}(e^{j\Omega}-e^{-j\Omega})$$
$$T(j\Omega)=e^{-j\Omega}2jsen(\Omega)$$
$$T(j\Omega)=e^{-j(\Omega-\frac{\pi}{2})}2sen(\Omega)$$
Fase: $e^{-j(\Omega-\frac{\pi}{2})}$

Módulo: $|2sen(\Omega)|$


```python
import numpy as np
import matplotlib.pyplot as plt

#Números de puntos de frecuncia
N = 1024
w = np.linspace(0, np.pi, N) #frecuencia angular de 0 a pi

#Funciones de transferencia
T_a = (1+np.exp(1j*w)+np.exp(2j*w)+np.exp(3j*w))/np.exp(3j*w)
T_b = (1+np.exp(1j*w)+np.exp(2j*w)+np.exp(3j*w)+np.exp(4j*w))/np.exp(4j*w)
T_c = 1-np.exp(-1j*w)
T_d = 1-np.exp(-2j*w)

# a. 
fase_a = np.angle(T_a)
modulo_a = np.abs(2*np.cos(w*3/2)+2*np.cos(w*1/2))

# b. 
fase_b = np.angle(T_b)
modulo_b = np.abs(1+(2*np.cos(2*w))+(2*np.cos(w)))
                  
# c.
fase_c = np.angle(T_c)
modulo_c = np.abs(2*np.sin(w/2))

# d. 
fase_d = np.angle(T_d)
modulo_d = np.abs(2*np.sin(w))

#%% GRÁFICOS

plt.figure()

plt.subplot(2,1,1)
plt.plot(w, modulo_a)
plt.title ('Respuesta en frecuencia a)')
plt.ylabel('Módulo [dB]')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(w, fase_a)
plt.ylabel('Fase [rad]')
plt.grid(True)

plt.figure()

plt.subplot(2,1,1)
plt.plot(w, modulo_b)
plt.title ('Respuesta en frecuencia b)')
plt.ylabel('Módulo [dB]')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(w, fase_b)
plt.ylabel('Fase [rad]')
plt.grid(True)

plt.figure()

plt.subplot(2,1,1)
plt.plot(w, modulo_c)
plt.title ('Respuesta en frecuencia c)')
plt.ylabel('Módulo [dB]')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(w, fase_c)
plt.ylabel('Fase [rad]')
plt.grid(True)

plt.figure()

plt.subplot(2,1,1)
plt.plot(w, modulo_d)
plt.title ('Respuesta en frecuencia d)')
plt.ylabel('Módulo [dB]')
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(w, fase_d)
plt.ylabel('Fase [rad]')
plt.grid(True)
```

Los sistemas a) y b) corresponden a filtros pasabajos basados en promedios móviles de 4 y 5 muestras respectivamente. Estos sistemas presentan:
- Módulo con ganancia máxima en bajas frecuencias y atenuación en altas.
- Dase con comportamiento lineal decreciente, lo que indica un retardo constante sin distorsión de fase.
- Cuanto mayor sea la longitud de promedio, mejor es la atenuación del ruido en alta frecuencia, pero también mayor el retardo.
El sistema c) es un filtro pasaaltos en donde:
- Módulo es nulo para bajas frecuencias y máximo para altas frecuencias.
- La fase muestra un comportamiento de avance de fase (por diferencia de muestras cercanas).
El sistema d) corresponde a un filtro pasabanda, en donde:
- Su respuesta en frecunecia presenta un pico en $\Omega=\frac{\pi}{2}$, atenuando tanto las bajas como las altas frecuencias.
